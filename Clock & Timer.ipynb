```python
#REAL TIME CLOCK

import tkinter as tk
from tkinter import font
import time

# Create the main window
window = tk.Tk()
window.title("Real-Time Clock")

# Create a font for the clock
font_style = font.Font(family="Helvetica", size=60)

# Create a label to display the clock
clock_label = tk.Label(window, font=font_style, bg="white", fg="black")
clock_label.pack(fill=tk.BOTH, expand=1)

# Create a function to update the clock every second
def update_clock():
    current_time = time.strftime("%I:%M:%S %p")
    clock_label.configure(text=current_time)
    window.after(1000, update_clock)

# Call the update_clock function to start the clock
update_clock()

# Run the main loop
window.mainloop()
```

```python
#TIMER

import tkinter as tk
import winsound

# Create the main window
window = tk.Tk()
window.title("Countdown Timer")

# Create a label to display the countdown timer
timer_label = tk.Label(window, text="Enter time in minutes and seconds:")
timer_label.pack()

# Create entry boxes for the user to enter the minutes and seconds
minutes_entry = tk.Entry(window)
minutes_entry.pack()
seconds_entry = tk.Entry(window)
seconds_entry.pack()

# Create a button to start the countdown
start_button = tk.Button(window, text="Start Countdown")
start_button.pack()

# Create a label to display the current time remaining
time_remaining_label = tk.Label(window)
time_remaining_label.pack()

# Create a function to start the countdown
def start_countdown():
    # Get the timer duration from the entry boxes
    minutes = int(minutes_entry.get())
    seconds = int(seconds_entry.get())
    duration = minutes * 60 + seconds

    # Set the time remaining to the duration
    time_remaining = duration

    # Update the time remaining label with the initial time
    time_remaining_label.configure(text=f"{minutes}:{seconds:02d}")

    # Create a function to update the time remaining
    def update_time():
        nonlocal time_remaining
        # Decrement the time remaining
        time_remaining -= 1

        # Calculate the minutes and seconds remaining
        minutes, seconds = divmod(time_remaining, 60)

        # Update the time remaining label
        time_remaining_label.configure(text=f"{minutes}:{seconds:02d}")

        # If the timer has not reached 0, call this function again
        # after 1000ms (1 second)
        if time_remaining > 0:
            window.after(1000, update_time)
        else:
            # Play a sound to indicate that the timer has finished
            winsound.PlaySound("sound.wav", winsound.SND_FILENAME)
    
    # Call the update_time function to start the countdown
    update_time()

# Set the command for the start button to the start_countdown function
start_button.configure(command=start_countdown)

# Run the main loop
window.mainloop()
```
